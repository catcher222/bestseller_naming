# 알라딘 영유아 부분 책 이름 크롤링

```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
```

```python
import time
```

```python
driver = webdriver.Chrome()
```

```python
for year in list(range(2015,2022)): #year 지정
    for month in list(range(1,13)): #month 지정
        for page in list([1,2,3,4]):
            # url에 년, 월, 페이지 바뀌는 곳에 반복되게 설정
            url = "https://www.aladin.co.kr/shop/common/wbest.aspx?BestType=MonthlyBest&BranchType=1&CID=13789&Year="+str(year)+"&Month="+str(month)+"&Week=1&page="+str(page)+"&cnt=1000&SortOrder=1" 
            driver.get(url)
            time.sleep(1)
            element = driver.find_elements_by_class_name("bo3")
            to_str = []
            to_df = pd.DataFrame()
            for i in list(range(0,50)):
                to_str.append(element[i].text)
            to_df = pd.DataFrame(to_str)
            to_df["rabel"] = 1
            book_name = book_name.append(to_df)   
            
            #page가 특수하게 설정되어 있어서 다른 문법을 적용
        for page in list([17,18,19,20]):
            url = "https://www.aladin.co.kr/shop/common/wbest.aspx?BestType=MonthlyBest&BranchType=1&CID=13789&Year="+str(year)+"&Month="+str(month)+"&Week=1&page="+str(page)+"&cnt=1000&SortOrder=1"
            driver.get(url)
            time.sleep(1)
            element = driver.find_elements_by_class_name("bo3")
            to_str = []
            to_df = pd.DataFrame()
            for i in list(range(0,50)):
                to_str.append(element[i].text)
            to_df = pd.DataFrame(to_str)
            to_df["rabel"] = 0 
            book_name = book_name.append(to_df)    
        
```
